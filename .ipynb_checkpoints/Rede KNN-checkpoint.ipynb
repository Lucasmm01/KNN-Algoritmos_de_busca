```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
import random
import math
```

```python
def gera_vertices(n):
    vertices = []
    random.seed(9)
    
    for i in range(n):
        x = random.randrange(0,n)
        y = random.randrange(0,n)
        vertices.append((x, y))
    return vertices
```

```python
def distancia_euclidiana(v1, v2):
    return math.sqrt(sum([(a - b) ** 2 for a, b in zip(v1, v2)]))
```

```python
def gera_arestas(lista_vertices, k):
    arestas = []
    for i, v1 in enumerate(lista_vertices):
        
        distancias = []
        for j, v2 in enumerate(lista_vertices):
            if i != j:
                dist = distancia_euclidiana(v1, v2)
                distancias.append((j, dist))
        
        #Ordena pela distancia em ordem crescente
        distancias.sort(key=lambda x: x[1])
        
        for j, dist in distancias[:k]:
            arestas.append((i, j, round(dist, 2)))
    return arestas
```

```python
def gera_rede_knn(n, k):
    lista_vertices = gera_vertices(n)
    lista_arestas = gera_arestas(lista_vertices, k)
    return lista_vertices, lista_arestas
```

```python
# gerar a rede
lista_vertices, lista_arestas = gera_rede_knn(100, 3)
```

### Busca em Largura

```python
def busca_profundidade(G, origem, destino):
    caminho = []
    visitados = set()
    pilha = [(origem, [origem])]
    
    while pilha:
        (vertice, caminho_atual) = pilha.pop()
        if vertice == destino:
            caminho = caminho_atual
            visitados.add(vertice)
            break
        if vertice not in visitados:
            visitados.add(vertice)
            for vizinho in G[vertice]:
                pilha.append((vizinho, caminho_atual + [vizinho]))
    
    return caminho, visitados
```

### Busca em profundidade

```python
from collections import deque

def bfs_shortest_path(vertices, edges, start, end):
    graph = {v: {} for v in vertices}  # dicionário para armazenar as informações do grafo
    for src, dest, weight in edges:
        graph[src][dest] = weight

    visited = set()  # conjunto para manter o controle dos nós já visitados
    queue = deque()  # fila para armazenar os nós a serem visitados
    queue.append(start)
    visited.add(start)
    prev = {v: None for v in vertices}  # dicionário para armazenar os predecessores

    while queue:
        node = queue.popleft()
        if node == end:
            break  # encontramos o nó de destino

        for neighbor, weight in graph[node].items():
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
                prev[neighbor] = node  # registra o predecessor

    # reconstrói o caminho mais curto
    path = []
    node = end
    while node is not None:
        path.append(node)
        node = prev[node]
    path.reverse()

    return path

vertices = ['A', 'B', 'C', 'D', 'E', 'F']
edges = [('A', 'B', 2), ('A', 'C', 3), ('B', 'D', 1), ('B', 'E', 4), ('C', 'F', 5), ('E', 'F', 1)]

path = bfs_shortest_path(vertices, edges, 'A', 'F')
print(path)
```
